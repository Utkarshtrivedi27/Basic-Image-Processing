```python
from skimage import io
from skimage.color import rgb2gray
```

```python
np_image1 = io.imread('sample1.jpg')
```

```python
import matplotlib.pyplot as plt
```

## Horizontal Flip

```python
plt.imshow(np_image1)
```

```python
image_flip1 = np_image1[:,::-1]      # horizontal flip
plt.imshow(image_flip1)
```

## Vertical Flip

```python
plt.imshow(np_image1)
```

```python
image_flip2 = np_image1[::-1]        # vertical flip
plt.imshow(image_flip2)
```

## Invert an Image

```python
plt.imshow(np_image1)
```

```python
image_invert = 255 - np_image1
plt.imshow(image_invert)
```

## Gray scale an Image

```python
np_image2 = io.imread('sample2.jpg')
plt.imshow(np_image2)
```

```python
image_gray_scale = rgb2gray(np_image2)
plt.imshow(image_gray_scale,cmap = 'gray',vmin= 0, vmax= 1)
```

## Image into B&W

```python
np_image3 = io.imread('sample2.jpg')
plt.imshow(np_image3)
```

```python
image_black_white = rgb2gray(np_image2)
image_black_white[image_black_white >= 0.85] = 1
image_black_white[image_black_white < 0.85] = 0

plt.imshow(image_black_white,cmap = 'gray', vmin = 0,vmax = 1)
```
